```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

In the following block of code we are using the pandas library to read our dataset from a csv file into a dataframe. I then use the 'df.head' method to print out the first few rows of data to quickly see the content of our dataset. After evaluating the first 5 rows of our dataset I came to the conclusion that the dataset is well-structured with clear features (YearsExperience) and a target variable (Salary) for building a LinearRegression model to estimate a salary based on years of experience. Finally, I select my x and y variables from the dataframe.

```python
df = pd.read_csv('Salary_Dataset.csv')

print(df.head())

x = df[['YearsExperience']] 
y = df['Salary']      
```

In the following block of code we will be training our model

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

In the following block of code we will be creating an instance of our LinearRegression model and fitting teh model with our x and y training

```python
model = LinearRegression()
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)

plt.scatter(x_test, y_test, color='blue', label='Actual data')
plt.plot(x_test, y_pred, color='red', label='Linear Regression')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

Evaluating our model

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2): {r2}')
```
